# َAnalysing of the effect of some features on rainfall in Australia

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv ("E:\Momeni-Machne Learning class momeni 21 01 1402\data science\datasets/weatherAUS.csv")
data
```

```python
df = pd.DataFrame (data)
df
```

```python
df.describe()
```

```python
df2 = pd.DataFrame(data, columns = ["Location", "WindGustDir","WindDir9am", "WindDir3pm"])
df2               
```

```python
df2.describe()
```

```python
df3 = df.dropna()
df4 = df2.dropna()
```

```python
df3
```

```python
plt.scatter (df3["MinTemp"], df3["Rainfall"])
plt.xlabel ("Min Temp")
plt.ylabel ('Rain')
```

```python
plt.scatter (df3["MaxTemp"], df3["Rainfall"])
plt.xlabel ("Max Temp")
plt.ylabel ('Rain')
```

```python
plt.scatter (df3["MinTemp"], df3["Rainfall"])
plt.scatter (df3["MaxTemp"], df3["Rainfall"], color='red')
plt.xlabel ("Max Temp")
plt.ylabel ('Rain')
```

```python
plt.figure (figsize = (10,6), dpi = 80)
plt.scatter (df3["Location"], df3["Rainfall"])
plt.xlabel ("Location")
plt.xticks (rotation = 90 )
plt.grid()
plt.ylabel ('Rain (mm)')
```

```python

plt.scatter (df3["WindGustSpeed"], df3["Rainfall"])
plt.xlabel ("WindGustSpeed")
plt.ylabel ('Rain (mm)')
```

```python
plt.figure (figsize = (10,6), dpi = 80)
plt.scatter (df3["WindSpeed9am"], df3["Rainfall"])
plt.scatter (df3["WindSpeed3pm"], df3["Rainfall"], color = 'red')
plt.xlabel ("WindSpeed")
plt.grid()
plt.ylabel ('Rain (mm)')
```
